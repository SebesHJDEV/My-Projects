
Check the Histogram fo Dataframe to see Data Distribution

```python
# Execute this to install matplotlib if you don't have it
#!pip install matplotlib
df.hist(figsize=(15, 10), bins= 1000)
```

The df.hist() method in Pandas is used to create histograms for each numerical column in a DataFrame. It visualizes the distribution of values within each column by dividing the data into intervals called "bins" and displaying the frequency or count of values falling within each bin.

Here's an explanation of the parameters commonly used with the df.hist() method:

figsize: This parameter allows you to specify the size of the figure (plot) in inches. It takes a tuple or list with two values representing the width and height of the figure. For example, figsize=(8, 6) sets the figure size to 8 inches wide and 6 inches tall.

bins: The bins parameter determines the number of bins or intervals to divide the data into. It controls the granularity of the histogram. Higher values of bins create more bins and provide more detail in the histogram, while lower values create fewer bins and provide a more generalized view. By default, the bins parameter is set to 10.

# Practical example how to use.

```python
# Create a variable to store the release dates of movies from 1980 to 2000
movies_1980_2000 = df[
    (df['release_date'].notnull()) & 
    (df['release_date'].dt.year >= 1980) & 
    (df['release_date'].dt.year <= 2000)
]
# Filter the DataFrame to include only movies released between 1980 and 2000,
# excluding any entries with null release dates.
# Now we will create a histogram using the filtered data.
movies_1980_2000.hist(figsize=(5, 5), bins=100, column="release_date")
```
